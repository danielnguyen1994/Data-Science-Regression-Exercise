```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets
from sklearn.linear_model import LogisticRegression
iris = datasets.load_iris()
```

```python
#Filtere zunächst das Dataset nach den 4 Eigenschaften 
X_p_width = iris["data"][:, 3:]  # petal width
X_p_length = iris["data"][:, 2:3] # petal length
X_s_width = iris["data"][:, 1:2]  # sepal width
X_s_length = iris["data"][:, 0:1] # sepal length

#Prüfe ob es sich um eine Iris-Virginica handelt
y = (iris["target"] == 2).astype(np.int)  # 1 if Iris-Virginica, else 0
```

```python
#Trainiere das Modell nach ihren 4 Eigenschaften 
log_reg_p_width = LogisticRegression(random_state=42)
log_reg_p_width.fit(X_p_width, y)

log_reg_p_length = LogisticRegression(random_state=42)
log_reg_p_length.fit(X_p_length, y)

log_reg_s_width = LogisticRegression(random_state=42)
log_reg_s_width.fit(X_s_width, y)

log_reg_s_length = LogisticRegression(random_state=42)
log_reg_s_length.fit(X_s_length, y)
```

```python
#Ermittele die Wahrscheinlichkeiten der Blumen mit einer passenden linspace()
X_new_p_width = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba_p_width = log_reg_p_width.predict_proba(X_new_p_width)

X_new_p_length = np.linspace(3, 6, 1000).reshape(-1, 1)
y_proba_p_length = log_reg_p_length.predict_proba(X_new_p_length)

X_new_s_width = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba_s_width = log_reg_s_width.predict_proba(X_new_s_width)

X_new_s_length = np.linspace(4, 8, 1000).reshape(-1, 1)
y_proba_s_length = log_reg_s_length.predict_proba(X_new_s_length)
```

```python
#Grpah von Petal Width
plt.plot(X_new_p_width, y_proba_p_width[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new_p_width, y_proba_p_width[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.xlabel("Petal Width", fontsize = 14)
```

```python
#Graph von Petal Length
plt.plot(X_new_p_length, y_proba_p_length[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new_p_length, y_proba_p_length[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.xlabel("Petal Length", fontsize = 14)
```

```python
#Graph von Sepal Width
plt.plot(X_new_s_width, y_proba_s_width[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new_s_width, y_proba_s_width[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.xlabel("Sepal Width", fontsize = 14)
```

```python
#Graph von Sepal Length
plt.plot(X_new_s_length, y_proba_s_length[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new_s_length, y_proba_s_length[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.xlabel("Sepal Length", fontsize = 14)
```

```python
#Ermittel den Schnittpunkt des Graphen von Petal Width
decision_boundary = X_new_p_width[y_proba_p_width[:, 1] >= 0.5][0]
decision_boundary
```

```python
#Ermittel den Schnittpunkt des Graphen von Petal Length
decision_boundary = X_new_p_length[y_proba_p_length[:, 1] >= 0.5][0]
decision_boundary
```

```python
#Ermittel den Schnittpunkt des Graphen von Sepal Width
decision_boundary = X_new_s_width[y_proba_s_width[:, 0] >= 0.5][0]
decision_boundary
```

```python
#Ermittel den Schnittpunkt des Graphen von Sepal Length
decision_boundary = X_new_s_length[y_proba_s_length[:, 1] >= 0.5][0]
decision_boundary
```
